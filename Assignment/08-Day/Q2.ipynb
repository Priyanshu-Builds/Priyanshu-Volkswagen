```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.arange(10)
plt.plot(x)
plt.title('Basic Line Plot')
plt.show()
```

```python
salaries=np.array([10,3,5,6,2,8,9,10,5,1])
expenses=np.array([2,4,1,5,2,6,4,6,2,1])
plt.plot(x,salaries,label='Salary')
plt.plot(x,expenses,label='Expenses')
plt.legend();plt.title('Salary vs Expenses');plt.show()
```

```python
plt.scatter(salaries,expenses);plt.title('Scatter');plt.show()
```

```python
plt.bar(x,salaries);plt.title('Bar');plt.show()
```

```python
plt.barh(x,expenses);plt.title('Horizontal Bar');plt.show()
```

```python
plt.fill_between(x,salaries);plt.title('Area');plt.show()
```

```python
plt.step(x,salaries);plt.title('Step');plt.show()
```

```python
plt.stem(x,salaries);plt.title('Stem');plt.show()
```
