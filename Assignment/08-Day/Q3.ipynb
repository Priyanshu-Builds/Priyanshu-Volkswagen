```python
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

```python
sal_data = pd.read_csv('Salary_Data.csv')
plt.figure(figsize=(8, 5))
plt.scatter(sal_data['YearsExperience'], sal_data['Salary'], color='blue', alpha=0.6)
plt.title('Experience vs Salary Correlation')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.grid(True)
plt.show()
```

```python
nba_df = pd.read_csv('nba.csv')
plt.figure(figsize=(8, 5))
plt.hist(nba_df['Age'].dropna(), bins=12, color='green', edgecolor='black')
plt.title('NBA Player Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
emp_df = pd.read_csv('empdata.csv')
dept_sal = emp_df.groupby('Department')['Salary'].sum().sort_values()
dept_sal.plot(kind='barh', color='orange')
plt.title('Total Salary Expenditure by Department')
plt.xlabel('Total Salary')
plt.show()
```
